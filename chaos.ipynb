```python
import random
import numpy as np
import matplotlib.pyplot as plt

point1 = np.array([0.,0.])
point2 = np.array([1.,0.])
point3 = np.array([0.5,np.sqrt(3.)/2.])

plt.figure()
point = point1
plt.plot(point1[0],point1[1],'x')
plt.plot(point2[0],point2[1],'x')
plt.plot(point3[0],point3[1],'x')
factor = 1./2.
while True :
  plt.draw()
  plt.pause(0.001)
  k = input('Enter 1, 2, or 3: ')
  #k = str(random.randint(1,4))
  if k=='1' :
    point = factor*point+(1-factor)*point1
  elif k=='2' :
    point = factor*point+(1-factor)*point2
  elif k=='3' :
    point = factor*point+(1.-factor)*point3
  plt.plot(point[0],point[1],'k.')
```
